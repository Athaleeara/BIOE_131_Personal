```python
#Opening text file
rjfile = open("romeoandjuliet.txt")


#Splittiing the lines into individual words and converting all text to uppercase
listed = []

punctuation = ['(', ')', '?', ':', ':', ',', '.', '!', '"', "'", '-','[',']', ';']


for line in rjfile:
    line = line.upper()
    for i in punctuation:
        line = line.replace(i,'')
    line = line.split()
    listed.append(line)
```

```python
dictionary = {}
def wordcounter(x):
    if x in dictionary:
        dictionary[x] = dictionary[x] + 1
    else:
        dictionary[x] = 1
        
for elm in listed:
    for subelm in elm:
        wordcounter(subelm)
```

```python
dict_list = []
for i,j in dictionary.items():
        dict_list.append([i,j])
def find_value(x):
    return x[1]
sorted_list = sorted(dict_list, key = find_value, reverse = True)
sorted_list
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pyplot as plt

x_data = range(1, len(sorted_list) + 1)
y_data = []
for elm in sorted_list:
    y_data.append(elm[1])

plt.plot(x_data, y_data, 'b')
plt.xlabel("Rank")
plt.ylabel("Word Count")
plt.title("Rank versus Word Count")
plt.show()
```

```python
import math
log_x = [math.log10(x) for x in x_data]
log_y = [math.log10(y) for y in y_data]

plt.plot(log_x, log_y, 'b')
plt.xlabel("Log(Rank)")
plt.ylabel("Log(Word Count)")
plt.title("Log(Rank) versus Log(Word Count)")
plt.show()
```
