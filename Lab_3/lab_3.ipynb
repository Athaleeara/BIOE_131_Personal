```python
#Read in text file "romeoandjuliet.txt"
rjfile = open("romeoandjuliet.txt")


#Split lines into individual words, convert text to uppercase, remove punctuation

listed = []
punctuation = ['(', ')', '?', ':', ':', ',', '.', '!', '"', "'", '-','[',']', ';']

for line in rjfile:
    line = line.upper()
    for i in punctuation:
        line = line.replace(i,'')
    line = line.split()
    listed.append(line)
```

```python
#Create dictionary with key = unique_ word and value = #_of_times_used

dictionary = {}

def wordcounter(x):
    #Count the number of times "x" is used in text and store this as the value of key "x"
    if x in dictionary:
        #if "x" is already in the dictionary increase vaule by 1
        dictionary[x] = dictionary[x] + 1
    else:
        #if "x" is not in the dictionary, store it with a value of 1
        dictionary[x] = 1
        
for elm in listed:
    #implement wordcounter function for each subelement of each elment of listed
    for subelm in elm:
        wordcounter(subelm)
```

```python
#Turn dictionary into a list
dict_list = []
for i,j in dictionary.items():
        #Add each item of dictionary to to dict_list as a list ["word", wordcount]
        dict_list.append([i,j])
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Plot a bar graph that shows the frequency each word appears, this 
# bar graph is not ordered

x_values = range(1, len(dict_list)+1)
y_values= [y[1] for y in dict_list]
plt.bar(x_values, y_values)
plt.ylabel("Word Count")
plt.title("Frequency the Words Appear Play (unordered)")
#Remove values from the x-axis since this plot is unordered
plt.xticks([])
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

#Sort the list of words used in the play by decreasing word count

def find_value(x):
    # Find the second element in a list that is defined as list = [key, value]
    return x[1]

#Create a sorted list of the words used in the play by decreasing word count
sorted_list = sorted(dict_list, key = find_value, reverse = True)


# Plot Rank versus Word Count 
x_data = range(1, len(sorted_list) + 1)
y_data = []
for elm in sorted_list:
    y_data.append(elm[1])

plt.bar(x_data, y_data)
plt.xlabel("Rank")
plt.ylabel("Word Count")
plt.title("Rank versus Word Count")
plt.show()
```

```python
import math

#Create new lists that are the log of each element in x_data and y_data
log_x = [math.log10(x) for x in x_data]
log_y = [math.log10(y) for y in y_data]

#Plot Log(Rank) versus Log(Word Count)
plt.plot(log_x, log_y)
plt.xlabel("Log (Rank)")
plt.ylabel("Log (Word Count)")
plt.title("Log (Rank) versus Log(Word Count)")
plt.show()
```

```python
# This plot shows a linear region where log(word count) and log(rank) 
# are directly related. Interpreting this and the previous plot, without the
# relationships, it appears that Romeo and Juliet follows Zipf's Law
# where the frequency of any word is inversely proportional to its rank.
```

```python
#Ten Most Frequent Words
for i in range(10):
    word = sorted_list[i][0]
    print("#", i+1, "most frequently used word:", word)
    
```
